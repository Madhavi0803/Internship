```python
import matplotlib.pyplot as plt
plt.hist(d1["age"])           # Histogram plot using matplotlib
plt.xlabel("age")
plt.ylabel("Count")
```

Here, I plotted a histogram for the numerical variable "age". The x-axis represents the numeric data and y-axis represents the respective count of the data.

Univariant Analysis of categorial data

```python
import seaborn as sns
sns.countplot(d1, x="gender")
```

 The above represents couunt plot. The count plot represents frequency distribution of categorical data. The above plot represents frequency distribution of variable "gender" of the given dataset.

```python
country_counts = d1['country'].value_counts()
plt.bar(country_counts.index, country_counts.values)
plt.title("countries Distribution")
plt.xlabel("country")
plt.ylabel("Count")
plt.show()
```

The above plot represents bar graph of categorial data of variable "country". x-axis represents country variable and y- axis represents count of the particular country.

```python
Gender = d1['gender'].value_counts()
plt.pie(Gender.values, labels=Gender.index, autopct='%1.1f%%')
plt.title("gender Distribution")
```

This represents pie chart of categorial variable "gender". It represent numerical proportions of a variable
